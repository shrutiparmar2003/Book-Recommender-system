```python
ratings=pd.read_csv(r"/content/Ratings.csv", low_memory=False)
```

```python
ratings.head()
```

```python
ratings.shape
```

```python
ratings.rename(columns={"User-ID":'user_id',
                      'Book-Rating':'no_of_ratings'},inplace=True)
```

```python
# Lets store users who had at least rated more than 200 books
x = ratings['user_id'].value_counts() > 200
```

```python
x[x].shape
```

```python
y= x[x].index
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python

ratings.head()
```

```python
# Joining ratings with books

ratings_with_books = ratings.merge(books, on='ISBN')
```

```python

ratings_with_books.head()
```

```python

ratings_with_books.shape
```

```python
number_rating = ratings_with_books.groupby('title')['no_of_ratings'].count().reset_index()
```

```python
number_rating.head()
```

```python
#merging num_rating table with ratings_with_books
final_rating = ratings_with_books.merge(number_rating, on='title')
```

```python
final_rating.head()
```

```python
final_rating.shape
```

```python
#books who got rated by equal to or more than 50 users

final_rating = final_rating[final_rating['no_of_ratings_y'] >= 50]
```

```python

final_rating.head()
```

```python

final_rating.shape
```

```python
# lets drop the duplicates
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python

final_rating.shape
```

```python
book_pivot= final_rating.pivot_table(columns='user_id',index='title', values='no_of_ratings_x')
```

```python
book_pivot.head()
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot.head()
```

```python
#scipy.sparse is used to store sparse or large matrices with zero values and CSR stands for compressed sparse matrix
#Training model starts from here
from scipy.sparse import csr_matrix
```

```python
book_sparse=csr_matrix(book_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
model= NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[0,:].values.reshape(1,-1), n_neighbors=6 )
```

```python
%%writefile app.py

import streamlit as st
import pickle

# Function to load artifacts
@st.cache_resource
def load_artifacts():
    with open('/content/model.pkl', 'rb') as f:
        model = pickle.load(f)
    with open('/content/book_name.pkl', 'rb') as f:
        books_name = pickle.load(f)
    with open('/content/final_rating.pkl', 'rb') as f:
        final_rating = pickle.load(f)
    with open('/content/book_pivot.pkl', 'rb') as f:
        book_pivot = pickle.load(f)

    return model, books_name, final_rating, book_pivot

# Load artifacts
model, books_name, final_rating, book_pivot = load_artifacts()

# Streamlit app
st.title("Book Recommender System")

selected_book_name = st.selectbox("Select a book to get recommendations", books_name)

if st.button('Recommend'):
    book_id = list(book_pivot.index).index(selected_book_name)
    distances, suggestions = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors=6)

    recommended_books = []
    for i in range(1, len(suggestions[0])):  # Start from 1 to skip the selected book itself
        recommended_books.append(books_name[suggestions[0][i]])

    st.write("Recommended Books:")
    for book in recommended_books:
        st.write(book)
```
